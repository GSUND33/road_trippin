```python
from bs4 import BeautifulSoup
import requests
import pymongo
import pandas as pd 
import numpy as np
```

```python
# URL of page to be scraped
url = 'https://en.wikipedia.org/wiki/List_of_bridge_failures'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')

print(soup.prettify())
#results = soup.find_all('table', class_='wikitable sortable jquery-tablesorter')
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_bridge_failures')
df_z = df[0]

df_fir = pd.read_html('https://en.wikipedia.org/wiki/List_of_bridge_failures')
df_fir = df[1]

df_sec = pd.read_html('https://en.wikipedia.org/wiki/List_of_bridge_failures')
df_sec = df[2]

df_thi = pd.read_html('https://en.wikipedia.org/wiki/List_of_bridge_failures')
df_thi = df[3]

df_fo = pd.read_html('https://en.wikipedia.org/wiki/List_of_bridge_failures')
df_fo = df[4]

total_frames = [df_z, df_fir, df_sec, df_thi, df_fo]
combined_dfs = pd.concat(total_frames)
combined_dfs.head()
```

```python
us_bridge_df = combined_dfs.loc[combined_dfs["Country"] == "United States", :]
us_bridge_df
```

```python
def ST(x):
    y=x.split(",")
    if len(y)>1:
        return y[1]
    else:
        return y[0]
```

```python
us_bridge_df["Location"]=us_bridge_df["Location"].loc[:].map(ST)
us_bridge_df.head()
```

```python
us_bridge_df = us_bridge_df.drop(['Comments', 'Unnamed: 10', 'Unnamed: 9'], axis=1)
```
