```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Wyoming_Road_Data.csv')
df.head()
```

```python
#select columns to keep by name
df = df[['State_Code','Route_ID','Route_Numb','Speed_Limi','IRI','AADT','Truck_NN','NHS','STRAHNET','Toll_Charg']]
df.rename(columns = {'Route_Numb':'Route_Number','Speed_Limi':'Speed_Limit','Truck_NN':'National_Truck_Network','Toll_Charg':'Toll_Charge'}, inplace = True)
df.head()
```

## Recode States

```python
#load state code definitions
state_codes = pd.read_csv('Recoding/recode_states.csv')
```

```python
df["State_Code"] = pd.to_numeric(df["State_Code"])
#merge state names to existing df
df = pd.merge(df, state_codes, left_on='State_Code', right_on='Code')
df.head()
```

```python
#Assign state code new values in 'Description'
df["State_Code"] = df["State_Code"].astype(str)
df["State_Code"] = df["Description"]
del df["Description"]
del df["Code"]
df.head()
```
