```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Tunnel_Details.csv',encoding = 'unicode_escape')
df.head()
```

```python
#select columns to keep by name
df = df[['Tunnel Number (I.1)','Tunnel Name (I.2)','State Code (I.3)','Route Type (I.9)',"Tunnel Portal's Latitude (I.13)",
         "Tunnel Portal's Longitude (I.14)",'Year Built (A.1)','Year Rehabilitated (A.2)','Number of Lanes (A.3)',
         'Annual Average Daily Traffic (A.4)','Annual Average Daily Truck Traffic (A.5)','Detour Length (A.7)','Toll (C.4)',
        'NHS Designation (C.5)','STRAHNET Designation (C.6)','Functional Classification (C.7)','Tunnel Length (G.1)',
        'Under Navigable Waterway (N.1)','Number of Bores (S.1)','Tunnel Shape (S.2)','Ground Conditions (S.4)']]

df.head()
```

```python
df1 = df.rename(columns = {'Tunnel Number (I.1)':'Tunnel_Number','Tunnel Name (I.2)':'Tunnel_Name','State Code (I.3)':'State_Code',
                     'Route Type (I.9)':'Route_Type',"Tunnel Portal's Latitude (I.13)":'Latitude',
                     "Tunnel Portal's Longitude (I.14)":'Longitude','Year Built (A.1)':'Year_Built',
                     'Year Rehabilitated (A.2)':'Year_Rehabilitated','Number of Lanes (A.3)':'Lanes',
                     'Annual Average Daily Traffic (A.4)':'AADT','Annual Average Daily Truck Traffic (A.5)':'AADTT',
                     'Detour Length (A.7)':'Detour Length','Toll (C.4)':'Toll','NHS Designation (C.5)':'NHS',
                     'STRAHNET Designation (C.6)':'STRAHNET','Functional Classification (C.7)':'Class',
                     'Tunnel Length (G.1)':'Length','Under Navigable Waterway (N.1)':'Under_Nav_Water',
                     'Number of Bores (S.1)':'Number_Bores','Tunnel Shape (S.2)':'Tunnel_Shape',
                     'Ground Conditions (S.4)':'Ground_Conditions'})
df1.head()
```

## Recode Values

### States

```python
#load state code definitions
state_codes = pd.read_csv('Recoding/recode_states.csv')
```

```python
df1["State_Code"] = pd.to_numeric(df1["State_Code"])
#merge state names to existing df
df = pd.merge(df1, state_codes, left_on='State_Code', right_on='Code')
df.head()
```

```python
#Assign state code new values in 'Description'
df["State_Code"] = df["State_Code"].astype(str)
df["State_Code"] = df["Description"]
del df["Description"]
del df["Code"]
df.head()
```

### Lat, Long

```python
#make long negative
df["Longitude"] = df["Longitude"]*-1
```

### Route_Type

```python
route_type_codes = pd.read_csv('Recoding/recode_tunnel_route.csv')

#recode " " to 0
df["Route_Type"] = df["Route_Type"].replace(to_replace =" ", 
                 value ="0")

df["Route_Type"] = pd.to_numeric(df["Route_Type"])
df = pd.merge(df, route_type_codes, left_on='Route_Type', right_on='Code')
df["Route_Type"] = df["Route_Type"].astype(str)
df["Route_Type"] = df["Description"]
del df["Description"]
del df["Code"]
```

### Class

```python
class_codes = pd.read_csv('Recoding/recode_tunnel_class.csv')
df["Class"] = df["Class"].replace(to_replace =" ", 
                 value ="0")
df["Class"] = pd.to_numeric(df["Class"])
df = pd.merge(df, class_codes, left_on='Class', right_on='Code')
df["Class"] = df["Class"].astype(str)
df["Class"] = df["Description"]
del df["Description"]
del df["Code"]
```

### Ground_Conditions

```python
ground_codes = pd.read_csv('Recoding/recode_tunnel_ground.csv')
df["Ground_Conditions"] = df["Ground_Conditions"].replace(to_replace =" ", 
                 value ="0")
df["Tunnel_Shape"] = pd.to_numeric(df["Ground_Conditions"])
df = pd.merge(df, ground_codes, left_on='Ground_Conditions', right_on='Code')
df["Ground_Conditions"] = df["Ground_Conditions"].astype(str)
df["Ground_Conditions"] = df["Description"]
del df["Description"]
del df["Code"]
```

### Tunnel_Shape

```python
shape_codes = pd.read_csv('Recoding/recode_tunnel_shape.csv')
df["Tunnel_Shape"] = df["Tunnel_Shape"].replace(to_replace =" ", 
                 value ="0")
df["Tunnel_Shape"] = pd.to_numeric(df["Tunnel_Shape"])
df = pd.merge(df, shape_codes, left_on='Tunnel_Shape', right_on='Code')
```

```python
df["Tunnel_Shape"] = df["Tunnel_Shape"].astype(str)
df["Tunnel_Shape"] = df["Description"]
del df["Description"]
del df["Code"]
```

## Tunnel Conditions

```python
cond_df = pd.read_csv('Tunnel_Condition.csv')
cond_df.head()
```

```python
cond_df['Condition State 3'] = pd.to_numeric(cond_df['Condition State 3'], errors='coerce')
cond_df['Condition State 4'] = pd.to_numeric(cond_df['Condition State 4'], errors='coerce')
cond_df['Count'] = np.where(((cond_df['Condition State 3']!=0) | (cond_df['Condition State 4']!=0)), 1, 0)
cond_df.head()
```

```python
cond_score = cond_df.groupby(['Tunnel Number'])['Count'].agg('sum')
cond_score = pd.DataFrame(cond_score)
cond_score= cond_score.reset_index()
cond_score.head()
```

## Merge Tunnel Details with Conditions

```python
df = pd.merge(df, cond_score, left_on='Tunnel_Number', right_on='Tunnel Number')
df = df.rename(columns = {'Count':'Condition'})
df = df.drop(columns=['Tunnel Number'])
df.head()
```
