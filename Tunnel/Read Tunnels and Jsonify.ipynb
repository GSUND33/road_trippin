```python
# SQL Alchemy
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()

# Config variables
from config import dbuser, dbpwd, dbhost, dbport, dbname

# Import Pandas
import pandas as pd
import pprint as pp

import json
```

```python
# Cloud MySQL Database Connection on AWS
engine = create_engine(f"mysql://{dbuser}:{dbpwd}@{dbhost}:{dbport}/{dbname}")
```

```python
# Create a remote database engine connection
conn = engine.connect()
```

```python
tunnels = pd.read_sql_query('select * from tunnels', con=engine)
tunnels.head()
```

```python
#select columns to keep by name
tunnels = tunnels[['Tunnel_Name','State_Code','Latitude','Longitude','Route_Type','Year_Built','Lanes','Length','Condition']]

tunnels.head()
```

```python
# json_data = roads.to_json(orient ='index')
tunnels_json = tunnels.to_dict('records')
print(json.dumps(tunnels_json, indent=4))
```

```python
with open('tunnels.txt', 'w') as outfile:
    json.dump(tunnels_json, outfile)
```
