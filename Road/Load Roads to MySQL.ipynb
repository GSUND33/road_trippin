```python
# SQL Alchemy
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()

# Config variables
from config import dbuser, dbpwd, dbhost, dbport, dbname

# Import Pandas
import pandas as pd
```

```python
# Cloud MySQL Database Connection on AWS
engine = create_engine(f"mysql://{dbuser}:{dbpwd}@{dbhost}:{dbport}/{dbname}")
```

```python
# Create a remote database engine connection
conn = engine.connect()
```

```python
csv_data = pd.read_csv('Road/All_Road_Data.csv')
csv_data.head()
```

```python
# Sum data up to state and location
del csv_data['System']
csv_data = csv_data.groupby(['State','Location'], as_index=False).sum()
```

```python
###compute ride quality summary score
#--------------
# <95  - Good (3)
# <170 - Acceptable (2)
# >170 - Bad (1)

csv_data['Ride_Quality'] = (3*(csv_data['<60']+csv_data['60-94']) +
                            2*(csv_data['95-119']+csv_data['120-144']+csv_data['145-170']) +
                            (csv_data['171-194']+csv_data['195-220']+csv_data['>220']))/(csv_data['<60']+csv_data['60-94']+
                                                                                         csv_data['95-119']+csv_data['120-144']+csv_data['145-170']+
                                                                                         csv_data['171-194']+csv_data['195-220']+csv_data['>220'])
```

```python
csv_data.to_sql(name='roads', if_exists='append', con=conn, chunksize=500, index=False)
```
