```python
# SQL Alchemy
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()

# Config variables
import sys
sys.path.append('..')
from config import remote_gwsis_dbuser, remote_gwsis_dbpwd, remote_db_host, remote_db_port, remote_gwsis_dbname

# Import Pandas
import pandas as pd
```

```python
# Cloud MySQL Database Connection on AWS
engine = create_engine(f"mysql://{remote_gwsis_dbuser}:{remote_gwsis_dbpwd}@{remote_db_host}:{remote_db_port}/{remote_gwsis_dbname}")
```

```python
# Create a remote database engine connection
conn = engine.connect()
```

```python
state_codes = pd.read_csv('geojson/state_code.csv')
state_codes.head()
```

```python
condition_data = pd.read_csv('road/All_Road_Data_Summary.csv')
condition_data.head()
```

```python
road_summary = condition_data.merge(state_codes, left_on=["State"], right_on=["State/District"])
road_summary = road_summary[["ID",
                                 "State",
                                 "Count_All",
                                 "Count_Good",
                                 "Count_Fair",
                                 "Count_Poor",
                                 "Area_All",
                                 "Area_Good",
                                 "Area_Fair",
                                 "Area_Poor",
                                 "Postal Code"
                                ]]
road_summary = road_summary.rename(columns={"Postal Code":"State_Abbreviation"})
road_summary.head()
```

```python
road_summary.to_sql(name='bridge_condition_summary', if_exists='replace', con=conn, chunksize=500, index=False)
```
