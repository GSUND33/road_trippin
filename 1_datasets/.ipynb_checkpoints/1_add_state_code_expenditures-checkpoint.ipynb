```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# assign federal dataset path to variable
csv_path = "geojson/state_code.csv"

# load csv
state_codes = pd.read_csv(csv_path, encoding="utf8")
state_codes.head(1)
```

### spending_state-local_2017dollars

```python
# assign federal dataset path to variable
csv_path = "expenditures/by_state/1_clean/spending_state-local_2017dollars.csv"

# load csv
state_totals = pd.read_csv(csv_path, encoding="utf8", low_memory=False)
state_totals.head(1)
```

```python
state_abbr = []

# for i in range(0,1): # use for testing
for i in range(0,len(state_totals)):
    
#     print(f"Processing index {i}...")
    
    df_name = state_totals['State'][i].lower()
    # print(df_name)
    
    if df_name == "united states":
        state_abbr.append("US")
        
    elif df_name == "dc":
            state_abbr.append("DC")
    
    else:
        for j in range(0,len(state_codes)):

            name = state_codes["State/District"][j].lower()
            code = state_codes["Postal Code"][j]

            if  df_name == name:
                state_abbr.append(code)

state_totals['State_Abbreviation'] = state_abbr
state_totals.head()
```

```python
state_totals = state_totals.reset_index()
state_totals = state_totals.rename(columns={"index":"ID"})
state_totals["ID"] = state_totals["ID"] + 1
state_totals.head()
```

```python
# export dataframe as csv
state_totals.to_csv("expenditures/by_state/spending_state-local_2017dollars.csv")
```
