```python
# SQL Alchemy
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()

# Config variables
import sys
sys.path.append('..')
from config import remote_gwsis_dbuser, remote_gwsis_dbpwd, remote_db_host, remote_db_port, remote_gwsis_dbname

# Import Pandas
import pandas as pd
```

```python
# Cloud MySQL Database Connection on AWS
engine = create_engine(f"mysql://{remote_gwsis_dbuser}:{remote_gwsis_dbpwd}@{remote_db_host}:{remote_db_port}/{remote_gwsis_dbname}")
```

```python
# Create a remote database engine connection
conn = engine.connect()
```

### spending_state-local_2017dollars

```python
csv_data = pd.read_csv('expenditures/by_state/spending_state-local_2017dollars.csv')
csv_data = csv_data.set_index("Unnamed: 0")
del csv_data.index.name
csv_data.head()
```

```python
csv_data.to_sql(name='spending_bystate_2017dollars', if_exists='replace', con=conn, chunksize=500, index=False)
```
