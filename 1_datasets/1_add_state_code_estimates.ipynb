```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# assign federal dataset path to variable
csv_path = "geojson/state_code.csv"

# load csv
state_codes = pd.read_csv(csv_path, encoding="utf8")
state_codes.head(1)
```

```python
# assign federal dataset path to variable
csv_path = "estimates/1_cleaned/sd2018_cost-good-rating.csv"

# load csv
estimate = pd.read_csv(csv_path, encoding="utf8", low_memory=False)
estimate.head(1)
```

```python
state_estimate = estimate.merge(state_codes, left_on=["State"], right_on=["State/District"])
state_estimate = state_estimate[["State","Count","Area",
                                 "Estimated Total Cost of Replacement",
                                 "Estimated Total Cost of Rehab",
                                 "Postal Code"]]
state_estimate = state_estimate.rename(columns={"Postal Code":"State_Abbreviation"})
state_estimate
```

```python
# export dataframe as a csv
state_estimate.to_csv("estimates/sd2018_cost-good-rating.csv", index=False)
```
