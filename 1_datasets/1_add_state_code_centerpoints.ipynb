```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# assign federal dataset path to variable
csv_path = "geojson/state_code.csv"

# load csv
state_codes = pd.read_csv(csv_path, encoding="utf8")
state_codes.head(1)
```

```python
# assign federal dataset path to variable
csv_path = "geojson/0_raw/state_centerpoints.csv"

# load csv
centerpoints = pd.read_csv(csv_path, encoding="utf8", low_memory=False)
centerpoints.head(1)
```

```python
state_centerpoints = centerpoints.merge(state_codes, left_on=["State"], right_on=["State/District"])
state_centerpoints = state_centerpoints[["State","Latitude","Longitude","Postal Code"]]
state_centerpoints = state_centerpoints.rename(columns={"Postal Code":"State_Abbreviation"})
state_centerpoints
```

```python
# export dataframe as a csv
state_centerpoints.to_csv("geojson/state_centerpoints.csv", index=False)
```
