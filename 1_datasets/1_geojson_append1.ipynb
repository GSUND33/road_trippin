```python
# import dependencies
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
# assign federal dataset path to variable
json_path = "geojson/gz_2010_us_040_00_5m.json"

# load csv
state_geo = pd.read_json(json_path, encoding="utf8")
state_geo.head()
```

```python
# assign features list to a variable
features = state_geo['features']

# normalize for ease of view
json_normalize(features)
```

```python
# assign bridge condition summary dataset path to variable
csv_path = "bridge/condition_by-hwy-sys_all_1.csv"

# load csv
bridge_df = pd.read_csv(csv_path)

# filter for 2018 data only; remove entries for Guam and Totals
data_2018 = bridge_df.loc[bridge_df["Year"] == 2018]
data_2018 = data_2018.loc[data_2018["State"] != "GUAM"]
data_2018 = data_2018.loc[data_2018["State"] != "TOTALS"]
data_2018
```

```python
data_2018["Tunnel_Percentage_Good"] = (data_2018["Tunnel_Percentage_Good"] *100).map("{:.2f}%".format)
data_2018["Tunnel_Percentage_Fair"] = (data_2018["Tunnel_Percentage_Fair"] *100).map("{:.2f}%".format)
data_2018["Tunnel_Percentage_Poor"] = (data_2018["Tunnel_Percentage_Poor"] *100).map("{:.2f}%".format)
data_2018["Tunnel_Percentage_Severe"] = (data_2018["Tunnel_Percentage_Severe"] *100).map("{:.2f}%".format)
data_2018["Area_Percentage_Good"] = (data_2018["Area_Percentage_Good"] *100).map("{:.2f}%".format)
data_2018["Area_Percentage_Fair"] = (data_2018["Area_Percentage_Fair"] *100).map("{:.2f}%".format)
data_2018["Area_Percentage_Poor"] = (data_2018["Area_Percentage_Poor"] *100).map("{:.2f}%".format)
data_2018["Percentage_Good"] = (data_2018["Percentage_Good"] *100).map("{:.2f}%".format)
data_2018["Percentage_Fair"] = (data_2018["Percentage_Fair"] *100).map("{:.2f}%".format)
data_2018["Percentage_Poor"] = (data_2018["Percentage_Poor"] *100).map("{:.2f}%".format)
data_2018["Road_NotReported_Percentage"] =(data_2018["Road_NotReported_Percentage"]).map("{:.2f}%".format)
data_2018["Road_Good_Percentage"] =(data_2018["Road_Good_Percentage"]).map("{:.2f}%".format)
data_2018["Road_Fair_Percentage"] =(data_2018["Road_Fair_Percentage"]).map("{:.2f}%".format)
data_2018["Road_Poor_Percentage"] =(data_2018["Road_Poor_Percentage"]).map("{:.2f}%".format)
data_2018.head()
```
