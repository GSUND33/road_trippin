```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# assign federal dataset path to variable
csv_path = "geojson/state_code.csv"

# load csv
state_codes = pd.read_csv(csv_path, encoding="utf8")
state_codes.head(1)
```

```python
# assign federal dataset path to variable
csv_path = "geojson/0_raw/state_boundaries.csv"

# load csv
boundaries = pd.read_csv(csv_path, encoding="utf8", low_memory=False)
boundaries.head(1)
```

```python
state_boundaries = boundaries.merge(state_codes, left_on=["State"], right_on=["State/District"])
state_boundaries = state_boundaries[["State","SouthBoundary","NorthBoundary","WestBoundary","EastBoundary","Postal Code"]]
state_boundaries = state_boundaries.rename(columns={"Postal Code":"State_Abbreviation"})
state_boundaries
```

```python
# export dataframe as a csv
state_boundaries.to_csv("geojson/state_boundaries.csv", index=False)
```
