```python
# import dependencies
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
# assign federal dataset path to variable
json_path = "geojs/gz_2010_us_040_00_5m.json"

# load csv
state_geo = pd.read_json(json_path, encoding="utf8")
state_geo.head(1)
```

```python
# assign features list to a variable
features = state_geo['features']

# normalize for ease of view
json_normalize(features)
```

```python
# assign bridge condition summary dataset path to variable
csv_path = "bridge/condition_by-hwy-sys_all.csv"

# load csv
bridge_df = pd.read_csv(csv_path)

# filter for 2018 data only; remove entries for Guam and Totals
data_2018 = bridge_df.loc[bridge_df["Year"] == 2018]
data_2018 = data_2018.loc[data_2018["State"] != "GUAM"]
data_2018 = data_2018.loc[data_2018["State"] != "TOTALS"]
data_2018
```

```python
type(data_2018["Count_All"])
```

```python
bridge_state = []

# for i in range(0,1): # use for testing
for i in range(0,len(data_2018)):
    condition = [
        ["bridge_count_all", data_2018.iloc[i,3]],
        ["bridge_count_good", data_2018.iloc[i,4]],
        ["bridge_count_fair", data_2018.iloc[i,5]],
        ["bridge_count_poor", data_2018.iloc[i,6]],
        ["bridge_area_all", data_2018.iloc[i,7]],
        ["bridge_area_good", data_2018.iloc[i,8]],
        ["bridge_area_fair", data_2018.iloc[i,9]],
        ["bridge_area_poor", data_2018.iloc[i,10]],
    ]
    if i == 0:
        bridge_state = [condition]
    else:
        bridge_state.append(condition)
```

```python
len(bridge_state)
```

```python
bridge_state
```

```python
for i in range(0,len(bridge_state)):
    
    feature = features[i]
    properties = bridge_state[i]
    
    for property in properties:
        feature["properties"][property[0]]=property[1]
        
json_normalize(features)
```
